```python
import torch
import pyro
import pyro.distributions as dist
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d = dist.DequantizedDistribution(dist.Binomial(4, 0.999))
x = torch.linspace(-1, 5, 1000)
plt.figure(figsize=(8, 3), dpi=300).patch.set_color('white')
plt.plot(x, d.log_prob(x).exp());
```

```python
plt.figure(figsize=(8, 3), dpi=300).patch.set_color('white')
plt.hist(d.sample((1000000,)), bins=200, range=(-1,5));
```
